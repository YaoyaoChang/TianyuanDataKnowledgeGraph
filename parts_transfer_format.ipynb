```python
import pandas as pd
```

```python
src_path = r"./data/part/MachineSystem_2.csv"
src = pd.read_csv(src_path)
src
```

```python
src['parentpart'] = src['MS_SystemName']
src
```

```python
src[src['MS_ID'] == 2]['MS_SystemName'].item()
```

```python
for index in src.index:
    if str(src.loc[index, 'MS_LevelOne_ID']) != 'nan':
        parent_id = int(src.loc[index, 'MS_LevelOne_ID'])
        parent = src[src['MS_ID'] == parent_id]['MS_SystemName'].item()
    else:
        parent = '挖掘机'
    src.loc[index,'parentpart'] = parent
#     print(index, parent_id, parent)

#     if index > 20:
#         break
src
```

```python
dst = src[['MS_SystemName','parentpart']]
dst = dst.rename(columns={'MS_SystemName':'part'})
dst
```

```python
dst.to_csv("./data/part/parts_from_ty.csv",quoting=1,sep=';',index=False,encoding='utf-16')
dst
```
